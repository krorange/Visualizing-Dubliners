```python
%matplotlib inline
import matplotlib.pyplot as plt

print (type(plt))

syllables = [134.98, 134.8, 141.8, 150]
tokens = [4265.53, 84713, 264185, 228626]
plt.scatter(syllables, tokens)
plt.xlabel('syllables')
plt.ylabel('tokens')
plt.title('syllables vs tokens')
```

```python
types = [1137.8, 13944, 30148, 57637]
plt.scatter(syllables, tokens, types)
plt.xlabel('syllables')
plt.ylabel('tokens')
plt.title('syllables vs tokens (size of bubble is types)')
```
